```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("student_scores.csv", sep=';')

# Display first rows
df.columns = ['study_hours', 'score']
print(df.head())
print(df.describe())
```

```python
# average score
mean_score = df["score"].mean()
print("Average Score:", mean_score)
```

```python
correlation = df["study_hours"].corr(df["score"])
print("Correlation between study hours and score:", correlation)
```

```python
# visualization
plt.scatter(df["study_hours"], df["score"])
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title("Study Hours vs Exam Score")
plt.show()
```
